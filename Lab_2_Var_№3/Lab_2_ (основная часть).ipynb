```python
#  
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

#    
D_values = [1.5, 2.0, 2.5]
alpha_values = [2.0, 3.0, 4.0]
beta_values = [1.0, 2.0, 3.0]

#  
t = np.linspace(0, 5, 500)

#  
lambda_vals = np.linspace(-5, 5, 500)

#   R(t)
def covariance_func(t, D, alpha, beta):
    return D * np.exp(-alpha * np.abs(t)) * np.cos(beta * t)

#  gamma(t) = R(0) - R(t)
def semivariogram(t, D, alpha, beta):
    return D - covariance_func(t, D, alpha, beta)

#   f(lambda)
def spectral_density(lambda_val, D, alpha, beta):
    term1 = D * alpha * (alpha**2 + lambda_val**2 + beta**2)
    term2 = (alpha**2 + (lambda_val - beta)**2) * (alpha**2 + (lambda_val + beta)**2)
    return term1 / (np.pi * term2)

#  
plt.figure(figsize=(18, 12))

#    D (alpha=2, beta=1)
plt.subplot(3, 3, 1)
for D in D_values:
    plt.plot(t, covariance_func(t, D, 2.0, 1.0), label=f'D={D}')
plt.title('R(t)   D (=2, =1)')
plt.xlabel('t')
plt.ylabel('R(t)')
plt.legend()

plt.subplot(3, 3, 2)
for D in D_values:
    plt.plot(t, semivariogram(t, D, 2.0, 1.0), label=f'D={D}')
plt.title('(t)   D (=2, =1)')
plt.xlabel('t')
plt.ylabel('(t)')
plt.legend()

plt.subplot(3, 3, 3)
for D in D_values:
    plt.plot(lambda_vals, spectral_density(lambda_vals, D, 2.0, 1.0), label=f'D={D}')
plt.title('f()   D (=2, =1)')
plt.xlabel('')
plt.ylabel('f()')
plt.legend()

#    alpha (D=1, beta=2)
plt.subplot(3, 3, 4)
for alpha in alpha_values:
    plt.plot(t, covariance_func(t, 1.0, alpha, 2.0), label=f'={alpha}')
plt.title('R(t)    (D=1, =2)')
plt.xlabel('t')
plt.ylabel('R(t)')
plt.legend()

plt.subplot(3, 3, 5)
for alpha in alpha_values:
    plt.plot(t, semivariogram(t, 1.0, alpha, 2.0), label=f'={alpha}')
plt.title('(t)    (D=1, =2)')
plt.xlabel('t')
plt.ylabel('(t)')
plt.legend()

plt.subplot(3, 3, 6)
for alpha in alpha_values:
    plt.plot(lambda_vals, spectral_density(lambda_vals, 1.0, alpha, 2.0), label=f'={alpha}')
plt.title('f()    (D=1, =2)')
plt.xlabel('')
plt.ylabel('f()')
plt.legend()

#    beta (D=1, alpha=1)
plt.subplot(3, 3, 7)
for beta in beta_values:
    plt.plot(t, covariance_func(t, 1.0, 1.0, beta), label=f'={beta}')
plt.title('R(t)    (D=1, =1)')
plt.xlabel('t')
plt.ylabel('R(t)')
plt.legend()

plt.subplot(3, 3, 8)
for beta in beta_values:
    plt.plot(t, semivariogram(t, 1.0, 1.0, beta), label=f'={beta}')
plt.title('(t)    (D=1, =1)')
plt.xlabel('t')
plt.ylabel('(t)')
plt.legend()

plt.subplot(3, 3, 9)
for beta in beta_values:
    plt.plot(lambda_vals, spectral_density(lambda_vals, 1.0, 1.0, beta), label=f'={beta}')
plt.title('f()    (D=1, =1)')
plt.xlabel('')
plt.ylabel('f()')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#  
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

#    
D_values = [1.5, 2.0, 2.5]
alpha_values = [2.0, 3.0, 4.0]

#  
t = np.linspace(0, 5, 500)

#  
lambda_vals = np.linspace(-5, 5, 500)

#   R(t)
def covariance_func(t, D, alpha):
    return D * np.exp(-alpha * np.abs(t)) * (1 + alpha * np.abs(t))

#  gamma(t) = R(0) - R(t)
def semivariogram(t, D, alpha):
    return D - covariance_func(t, D, alpha)

#   f(lambda)
def spectral_density(lambda_val, D, alpha):
    return (D / np.pi) * (2 * alpha**3) / (alpha**2 + lambda_val**2)**2

#  
plt.figure(figsize=(18, 12))

#    D (alpha=3)
plt.subplot(3, 3, 1)
for D in D_values:
    plt.plot(t, covariance_func(t, D, 3.0), label=f'D={D}')
plt.title('R(t)   D (=3)')
plt.xlabel('t')
plt.ylabel('R(t)')
plt.legend()

plt.subplot(3, 3, 2)
for D in D_values:
    plt.plot(t, semivariogram(t, D, 3.0), label=f'D={D}')
plt.title('(t)   D (=3)')
plt.xlabel('t')
plt.ylabel('(t)')
plt.legend()

plt.subplot(3, 3, 3)
for D in D_values:
    plt.plot(lambda_vals, spectral_density(lambda_vals, D, 3.0), label=f'D={D}')
plt.title('f()   D (=3)')
plt.xlabel('')
plt.ylabel('f()')
plt.legend()

#    alpha ( D=1)
plt.subplot(3, 3, 4)
for alpha in alpha_values:
    plt.plot(t, covariance_func(t, 1.0, alpha), label=f'={alpha}')
plt.title('R(t)    (D=1)')
plt.xlabel('t')
plt.ylabel('R(t)')
plt.legend()

plt.subplot(3, 3, 5)
for alpha in alpha_values:
    plt.plot(t, semivariogram(t, 1.0, alpha), label=f'={alpha}')
plt.title('(t)    (D=1)')
plt.xlabel('t')
plt.ylabel('(t)')
plt.legend()

plt.subplot(3, 3, 6)
for alpha in alpha_values:
    plt.plot(lambda_vals, spectral_density(lambda_vals, 1.0, alpha), label=f'={alpha}')
plt.title('f()    (D=1)')
plt.xlabel('')
plt.ylabel('f()')
plt.legend()

plt.tight_layout()
plt.show()
```
